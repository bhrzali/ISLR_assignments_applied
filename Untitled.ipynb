```r
library(MASS)
summary(Boston)
```

```r
Boston2 = Boston
coef_simple = c()
Boston2$chas = as.factor(Boston$chas)
for (p in 2:length(Boston2)){
    pred = Boston[,p]
    print(colnames(Boston2)[p])
    lm.fit = lm(crim~pred,data=Boston)
    coef_simple = c(coef_simple, coef(lm.fit)[-1])
    print(summary(lm.fit))
}
```

```r
lm.fit2 = lm(crim~.,data=Boston)
summary(lm.fit2)
plot(coef_simple,coef(lm.fit2)[-1])
```
